```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
from itertools import groupby

from utils import *
from visualize import *

raw_csv = 'data.csv'
```

```python
data = pd.read_csv(raw_csv, skiprows=[0,1,2,3], names=(['frame'] + HEADERS)).set_index('frame')
data
```

```python
headers1 = [i[:-4] for i in CCOLS for j in '1234']
headers2 = [j for i in CCOLS for j in ('#of0', 'cons', '1stQ','10thQ')]
```

```python
values = []
for col_name in CCOLS:
    print(col_name[:-4]+'\n------')
    col = data[col_name].fillna(0)
    
    zeros = (col==0)
    nums = zeros.sum()
    consecutive = max([len(list(g)) for k, g in groupby(zeros) if k]) if nums > 0 else 0
    first = str(round(col.quantile(q=0.01), 2))
    tenth = str(round(col.quantile(q=0.10), 2))
    
    print(f"# of 0s        : {nums}")
    print(f"consecutive 0s : {consecutive}")
    print(f"1st Q          : {first}")
    print(f"10th Q         : {tenth} \n")
    values += [nums, consecutive, first, tenth]
```

```python
pd.DataFrame(np.vstack((headers1, headers2, values)))
```
