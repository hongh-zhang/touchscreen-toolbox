```python
from scipy import signal

def median_filter(df, length):
    """
    Apply median filter to each column of the give [df],
    not in-place
    """
    df.apply(signal.medfilt, args=(length,), axis=0)
```

```python
import cv2

cap = cv2.VideoCapture('1_clip.mp4')
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
writer = cv2.VideoWriter('out.mp4', fourcc, 25, (640, 480))

for (idx, row) in coords.iterrows():
    
    ret, frame = cap.read()
    if ret:
        frame = cv2.circle(frame, (round(row[0]), round(row[1])), 5, (255,0,0), -1)
        writer.write(frame)
    else:
        break
        
cap.release()
writer.release()
```
